## Assessing and Cleaning NY AIRBNB 2019 dataset

This dataset is about airbnb listing activity of NewYork city of year 2019. This dataset is public and the original source can be found on this https://insideairbnb.com
The dataset includes multiple columns provided with information about hosts, geographical availability, necessary metrics. These metrics can be used to make predictions and draw conclusions. 

 
  Summary To be continued.....

### Column Description:
- `id` = listing id on Airbnb.
- `name` = name of listing.
- `host_id` = Identifies a specific user's profile on the Airbnb platform.
- `host_name` = host name or display name of Airbnb account.
- `neighbourhood_group` = community location.
- `neighbourhood` = Guests can search for specific neighborhoods.
- `latitude` = latitude coordinates of location.
- `longitude` = longitude coordinates of location.
- `room_type` = listing space type.
- `price` = listing price in dollar.
- `minimum_nights` = amount of amount of minimum nights for listing.
- `number_of_reviews` = number of reviews by the guest.
- `last_review` = most latest review by the guest. 
- `reviews_per_month` = number of reviews per month by the guest.
- `calculated_host_listings_count` = amount of listing per host. 
- `availability_365` = number of days when listing is available for booking
